### 2. Analyse de la complétude des données

```python
# -------------------------------------------
# 1. Liste des fichiers-candidates à analyser
# -------------------------------------------
fichiers = [
    "Access_to_electricity.csv",
    "Agricultural_land.csv",
    "Annual_inflation_consumer_prices.csv",
    "Electric_power_consumption.csv",
    "Fixed_broadband_per_100_people.csv",
    "GDP_growth_annual_part.csv",
    "GDP_per_capita.csv",
    "Imports_of_goods.csv",
    "Lead_time_to_import_days.csv",
    "Literacy_rate_adult_total_ages_15_and_above.csv",
    "Logistics_performance_index.csv",
    "Merchandise_trade_part_of_GDP.csv",
    "Population_65_and_above_part_of_total.csv",
    "Population_density.csv",
    "Populations_15_64_part_of_total.csv",
    "Undernourishment_part_of_population.csv",
    "Urban_population_part.csv"
]

# Colonnes annuelles
colonnes_annees = [str(a) for a in range(2000, 2019)]

# Résultats finaux
resultats = []


# -----------------------------------------------------------------------
# 2. Boucle sur tous les fichiers pour analyser le nombre de lignes vides
# -----------------------------------------------------------------------
for file in fichiers:

    try:
        # Lecture du fichier
        df = pd.read_csv(file, skiprows=4)

        # Sélection des colonnes existantes (pour éviter les erreurs)
        colonnes_existantes = [c for c in colonnes_annees if c in df.columns]
        colonnes_a_garder = ["Country Code"] + colonnes_existantes

        df = df[colonnes_a_garder]

        # Nombre total de lignes
        nb_lignes = len(df)

        # Lignes où TOUTES les années sont NaN
        nb_lignes_tout_nan = df[colonnes_existantes].isna().all(axis=1).sum()

        # Lignes où AU MOINS un NaN existe
        nb_lignes_avec_nan = df[colonnes_existantes].isna().any(axis=1).sum()

        # Stocker le résultat
        resultats.append({
            "Fichier": file,
            "Total lignes": nb_lignes,
            "Toutes années NaN": nb_lignes_tout_nan,
            "Au moins 1 NaN": nb_lignes_avec_nan,
            "% lignes propres": round(100 * (nb_lignes - nb_lignes_avec_nan) / nb_lignes, 1)
        })

    except Exception as e:
        resultats.append({
            "Fichier": file,
            "Erreur": str(e)
        })

# --------------------------------------------
# 3. Afficher le résumé pour tous les fichiers
# --------------------------------------------
df_resultats = pd.DataFrame(resultats)
print(df_resultats)
```

Indicateurs sélectionnés pour l'analyse selon la méthode PESTEL :

* Social (S) : Poultry availability 2017 (kg/person/year) — mesure la consommation réelle de viande de poulet par habitant dans le pays sur une année. Pourquoi est-ce important pour l'analyse ? Cela montre l'habitude prise par la population de consommer du poulet, c'est-à-dire une valeur élevée (> 20 kg/an) = un marché « mature », les gens savent cuisiner et apprécier le produit, et une valeur faible (< 5 kg/an) = une adaptation culturelle, du marketing et une éducation des consommateurs sont nécessaires. En d'autres termes, c'est un indicateur de la facilité d'entrée : il est plus facile de vendre là où l'on mange déjà. Pour la classification : division des pays en « chicken lovers » et « low consumers »

* Political (P) + Economic (E) : Poultry imports per capita 2017 (kg/person) — mesure le volume de poulet importé par habitant (indicateur normalisé). Important pour l'analyse : montre l'ouverture politique, par exemple l'expérience existante en matière d'importation sous forme de canaux de distribution et de procédures douanières bien établies. Prouve également la demande, car lorsque les importations sont élevées, la production locale ne couvre pas les besoins. Pour la classification : identifie les pays fortement dépendants des importations, qui constituent les meilleures cibles pour l'exportation


* Economic (E) + Environmental (En) : Poultry self-sufficiency 2017 (ratio) — mesure le coefficient d'autosuffisance = Production / Consommation totale. Si >1 → le pays produit plus qu'il ne consomme (exportateur),  si <1 → déficit, importations nécessaires, si ~0 → presque pas de production locale. Pourquoi est-ce important pour l'analyse : permet d'évaluer la concurrence, avec une auto-suffisance élevée (>0,9) = producteurs locaux forts, difficile d'entrer sur le marché. Permet d'identifier le déficit : faible ratio (<0,5) = forte dépendance vis-à-vis des importations → bonne opportunité. En tant qu'aspect environnemental : montre le développement de l'aviculture et du secteur agricole. Pour la classification : divise en « marchés autosuffisants » et « marchés dépendants des importations » 

* Economic (E) : Population total 2017 (millions) — mesure la population totale du pays en millions. Pourquoi est-ce important pour l'analyse ? Cela montre la taille du marché en termes absolus, même avec une consommation moyenne, 100 millions d'habitants → un volume de ventes énorme. Il permet d'évaluer l'efficacité logistique : un grand marché = possibilité de construire des entrepôts régionaux, de réduire les coûts unitaires. Pour la classification : distingue les « marchés géants » des « marchés de niche ».

Il démontre le dividende démographique : une croissance rapide (> 2 %) signifie souvent une population jeune → une consommation plus importante de protéines. Il montre l'attrait des investissements : une dynamique positive → il vaut la peine d'investir maintenant pour répondre à la demande future. Pour le regroupement : il divise les marchés en « marchés en croissance » et « marchés stagnants »

* Economic (E) : GDP per capita 2017 ($) — mesure le PIB par habitant en dollars américains (courants). Pourquoi est-ce important pour l'analyse ? Permet d'évaluer la demande solvable, car le poulet biologique coûte 1,5 à 3 fois plus cher que le poulet conventionnel. Effet seuil : lorsque le PIB par habitant est supérieur à 10 000 dollars, la demande de produits haut de gamme augmente considérablement. Également, une croissance moyenne élevée (> 6 %) = marché en développement, classe moyenne en expansion. Pour la classification : identifie les « économies à croissance rapide » par rapport aux « économies stagnantes »

* Political (P) + Legal (L) : Goods import 2017 (% of GDP) — mesure la part des importations de marchandises dans le PIB du pays. Pourquoi est-ce important pour l'analyse ? Environnement réglementaire : un faible pourcentage correspond souvent à du protectionnisme, des douanes complexes, des quotas. ### 2. Filtrage afin de ne conserver que les colonnes nécessaires à l'analyse

```python
# Téléchargement et traitement de 12 fichiers de la Banque mondiale

# Créer une liste des années de 2000 à 2018 sous forme de texte ("2000", "2001", etc.)
colonnes_annees = [str(a) for a in range(2000, 2019)]

# Boucle : traiter chaque fichier de la liste fichiers_worldbank
for file_name in fichiers_worldbank:
    # Construire le chemin complet du fichier (dossier + nom)
    chemin_complet = os.path.join(DATA_PATH_WB, file_name)
    
    # Lire le fichier CSV (skiprows=4 ignore les 4 premières lignes de métadonnées)
    df_raw = pd.read_csv(chemin_complet, skiprows=4)
    
    # Vérifier si c'est le fichier GDP growth annual (celui qui contient plusieurs années)
    if file_name == "GDP growth annual.csv":
        # Pour GDP growth : garder Country Name, Country Code et TOUTES les années
        colonnes_a_garder = ["Country Name", "Country Code"] + colonnes_annees
        # Copier seulement les colonnes nécessaires
        df_filtre = df_raw[colonnes_a_garder].copy()
        # Calculer la moyenne de croissance sur toutes les années (2000-2018)
        # axis=1 signifie calculer la moyenne horizontalement (pour chaque pays)
        df_filtre["GDP growth avg 2000-2018 (%)"] = df_filtre[colonnes_annees].mean(axis=1).round(2)
        # Garder seulement 3 colonnes : Country Name, Country Code et la nouvelle moyenne
        df_final = df_filtre[["Country Name", "Country Code", "GDP growth avg 2000-2018 (%)"]].copy()
    
    # Pour tous les autres fichiers (11 fichiers avec données 2017)
    else:
        # Garder seulement Country Name, Country Code et l'année 2017
        colonnes_a_garder = ["Country Name", "Country Code", "2017"]
        # Copier seulement ces 3 colonnes
        df_filtre = df_raw[colonnes_a_garder].copy()
        # Extraire le nom du fichier sans l'extension ".csv"
        # Exemple : "GDP per capita.csv" → "GDP per capita"
        metric_name = file_name.replace(".csv", "")
        # Créer un nouveau nom de colonne descriptif
        # Exemple : "GDP per capita 2017"
        new_col_name = f"{metric_name} 2017"
        # Renommer la colonne "2017" en nom descriptif
        df_filtre.rename(columns={"2017": new_col_name}, inplace=True)
        # Arrondir toutes les valeurs à 2 décimales
        df_filtre[new_col_name] = df_filtre[new_col_name].round(2)
        # Garder seulement 3 colonnes : Country Name, Country Code et la métrique 2017
        df_final = df_filtre[["Country Name", "Country Code", new_col_name]].copy()
    
    # Créer un nom pour la variable DataFrame à partir du nom du fichier
    # Exemple : "GDP per capita.csv" → "df_GDP_per_capita"
    df_name = "df_" + file_name.replace(".csv", "").replace(" ", "_")
    # Enregistrer le DataFrame dans les variables globales avec ce nom
    # Cela crée une variable que nous pouvons utiliser plus tard
    globals()[df_name] = df_final

# Afficher un message de confirmation
print("✓ Tous les 12 fichiers World Bank ont été chargés et traités avec succès")
```

### 3. Analyse de la qualité des données (détection des valeurs manquantes)

```python
# ========================================================================
# Analyse de la complétude des données pour les 12 fichiers World Bank
# ========================================================================

# Liste des noms de nos 12 DataFrames à analyser
dfs_a_analyser = [
    "df_GDP_growth_annual",
    "df_GDP_per_capita",
    "df_Goods_import",
    "df_Electricity_acces",
    "df_Agricultural_land",
    "df_Electric_power_consumption",
    "df_Merchandise_trade_part_of_GDP",
    "df_Population_65_and_above_part_of_total",
    "df_Population_density",
    "df_Urban_population_part",
    "df_Populations_15_64_part_of_total",
    "df_Undernourishment_part_of_population"
]

# Analyse de chaque DataFrame
resultats = []                                                # Liste vide pour stocker les résultats
for df_name in dfs_a_analyser:                                # Boucle sur chaque DataFrame
    df_actuel = globals()[df_name]                            # Récupère le DataFrame par son nom
    nb_pays_total = len(df_actuel)                            # Nombre total de pays (lignes)
    colonne_donnees = df_actuel.columns[2]                    # Nom de la colonne de données (3ème colonne)
    nb_nan = df_actuel[colonne_donnees].isna().sum()          # Nombre de valeurs NaN dans cette colonne
    pct_completude = round(((nb_pays_total - nb_nan) / nb_pays_total) * 100, 1) # % de complétude
    
    resultats.append({
        "Métrique": colonne_donnees,                          # Nom de la métrique
        "Total pays": nb_pays_total,                          # Nombre total de pays
        "Valeurs valides": nb_pays_total - nb_nan,            # Nombre de valeurs valides
        "% complétude": pct_completude                        # Pourcentage de complétude
    })

df_completude = pd.DataFrame(resultats)                       # Convertir en DataFrame pour affichage
print("\nANALYSE DE LA QUALITÉ DES DONNÉES - World Bank\n")   # Titre de la section
print(df_completude.to_string(index=False))                   # Afficher sans les index

# Statistiques globales
print(f"\nComplétude moyenne : {df_completude['% complétude'].mean():.1f}%") # Moyenne des % de complétude
```

```python
# ========================================================================
# IDENTIFICATION DES PAYS AVEC VALEURS MANQUANTES
# ========================================================================
# Cette section montre QUELS pays ont des NaN pour chaque indicateur
# Cela aide à décider s'il faut supprimer ces pays ou non

print("\n" + "="*70)
print("LISTE DES PAYS AVEC VALEURS MANQUANTES (NaN)")
print("="*70)

# Boucle FOR : parcourt à nouveau chaque DataFrame
for df_name in dfs_a_analyser:
    # Récupération du DataFrame actuel
    df_actuel = globals()[df_name]
    
    # Récupération du nom de la colonne de données (3ème colonne)
    colonne_donnees = df_actuel.columns[2]
    
    # Filtrer pour obtenir seulement les lignes où la colonne de données est NaN
    # df_actuel[colonne_donnees].isna() : retourne True/False pour chaque ligne
    # df_actuel[condition] : garde seulement les lignes où condition = True
    pays_avec_nan = df_actuel[df_actuel[colonne_donnees].isna()]
    
    # Vérifier s'il y a des pays avec NaN
    if len(pays_avec_nan) > 0:
        print(f"\n🔍 {colonne_donnees}")
        print(f"   Nombre de pays sans données : {len(pays_avec_nan)}")
        print("   Liste des pays concernés :")
        
        # Afficher les noms des pays (colonne "Country Name")
        # .values : récupère les valeurs de la colonne sous forme de liste
        # enumerate() : donne un numéro à chaque élément (commence à 1 avec start=1)
        for i, pays in enumerate(pays_avec_nan["Country Name"].values, start=1):
            print(f"      {i}. {pays}")
    else:
        # Si aucun NaN trouvé
        print(f"\n✓ {colonne_donnees} : Aucune valeur manquante")

print("\n✓ Analyse des pays avec valeurs manquantes terminée")
```

### 4. Traitement des valeurs NaN

```python
# ========================================================================
# Suppression des pays avec valeurs manquantes (NaN)
# ========================================================================
# Stratégie : supprimer toutes les lignes (pays) qui ont au moins un NaN
# dans l'un des 4 indicateurs, car le clustering ne peut pas traiter les NaN

# Liste des 4 DataFrames à nettoyer
# Ce sont les mêmes que dans les sections précédentes
dfs_a_nettoyer = [
    "df_GDP_growth_annual",       
    "df_GDP_per_capita",         
    "df_Goods_import",            
    "df_Electricity_acces",
    "df_Agricultural_land",
    "df_Electric_power_consumption",
    "df_Merchandise_trade_part_of_GDP",
    "df_Population_65_and_above_part_of_total",
    "df_Population_density",
    "df_Urban_population_part",
    "df_Populations_15_64_part_of_total",
    "df_Undernourishment_part_of_population"      
]

print("="*70)
print("NETTOYAGE DES DONNÉES - Suppression des valeurs NaN")
print("="*70)

# Parcourir chaque DataFrame
for df_name in dfs_a_nettoyer:
    # Récupération du DataFrame actuel
    # globals()[df_name] : accède à la variable globale par son nom
    df_actuel = globals()[df_name]
    
    # Compter le nombre de lignes AVANT la suppression
    nb_lignes_avant = len(df_actuel)
    
    # Récupération du nom de la colonne de données (3ème colonne, index 2)
    colonne_donnees = df_actuel.columns[2]
    
    # SUPPRESSION des lignes avec NaN dans la colonne de données
    # .dropna() : supprime toutes les lignes contenant au moins un NaN
    # subset=[colonne_donnees] : ne vérifie que cette colonne spécifique
    # inplace=False : retourne un nouveau DataFrame (ne modifie pas l'original)
    df_nettoye = df_actuel.dropna(subset=[colonne_donnees])
    
    # Compter le nombre de lignes APRÈS la suppression
    nb_lignes_apres = len(df_nettoye)
    
    # Calculer combien de lignes ont été supprimées
    nb_lignes_supprimees = nb_lignes_avant - nb_lignes_apres
    
    # IMPORTANT : remplacer l'ancien DataFrame par le nouveau (nettoyé)
    # globals()[df_name] = ... : met à jour la variable globale
    # Maintenant df_GDP_per_capita (par exemple) contient le DataFrame nettoyé
    globals()[df_name] = df_nettoye
    
    # Affichage des résultats du nettoyage
    print(f"\n🧹 {df_name}")
    print(f"   Métrique : {colonne_donnees}")
    print(f"   Pays avant nettoyage : {nb_lignes_avant}")
    print(f"   Pays après nettoyage : {nb_lignes_apres}")
    print(f"   Pays supprimés : {nb_lignes_supprimees}")

print("\n" + "="*70)
print("✓ Nettoyage terminé - Les 4 DataFrames ne contiennent plus de NaN")
print("="*70)
```

### 6. Fusion des 4 DataFrames World Bank en un seul df_worldbank

```python
# ========================================================================
# Fusion des 4 DataFrames World Bank en un seul DataFrame
# ========================================================================
# But : combiner les 4 indicateurs en un seul tableau avec une ligne par pays
# Méthode : INNER JOIN sur "Country Code" et "Country Name"
# Résultat : seuls les pays présents dans TOUS les 4 DataFrames seront conservés

import functools

# Liste des 4 DataFrames à fusionner

dfs_a_fusionner = [
    df_GDP_growth_annual,                       # 3 colonnes : Country Name, Code, GDP growth avg 2000-2018 (%)
    df_GDP_per_capita,                          # 3 colonnes : Country Name, Code, GDP per capita 2017, current US$
    df_Goods_import,                            # 3 colonnes : Country Name, Code, Goods import 2017, % of GDP
    df_Electricity_acces,                       # 3 colonnes : Country Name, Code, Electricity acces 2017, % of population
    df_Agricultural_land,                       # 3 colonnes : Country Name, Code, Agricultural land 2017, % of land area
    df_Electric_power_consumption,              # 3 colonnes : Country Name, Code, Electric power consumption 2017, kWh per capita
    df_Merchandise_trade_part_of_GDP,           # 3 colonnes : Country Name, Code, Merchandise trade 2017, % of GDP
    df_Population_65_and_above_part_of_total,   # 3 colonnes : Country Name, Code, Population 65+ 2017, % of total
    df_Population_density,                      # 3 colonnes : Country Name, Code, Population density 2017, per sq. km
    df_Urban_population_part,                   # 3 colonnes : Country Name, Code, Urban population 2017, % of total
    df_Populations_15_64_part_of_total,         # 3 colonnes : Country Name, Code, Population 15-64 2017, % of total
    df_Undernourishment_part_of_population      # 3 colonnes : Country Name, Code, Undernourishment 2017, % of population
]

# Clés de jointure : les 2 premières colonnes communes à tous les DataFrames
# Country Code : code ISO du pays (ex: "FRA", "USA")
# Country Name : nom du pays (ex: "France", "United States")
cles_jointure = ["Country Code", "Country Name"]

# FUSION avec functools.reduce()
# functools.reduce() : applique une fonction de manière cumulative sur une liste
# Étapes :
#   1. Fusionne df1 et df2 → résultat1
#   2. Fusionne résultat1 et df3 → résultat2
#   3. Fusionne résultat2 et df4 → résultat final
#
# pd.merge() avec how="inner" :
#   - Garde SEULEMENT les pays présents dans les DEUX DataFrames
#   - Si un pays manque dans un DataFrame, il est supprimé du résultat
#   - Cela garantit que tous les pays ont des données pour TOUS les indicateurs
df_worldbank = functools.reduce(
    lambda left, right: pd.merge(left, right, on=cles_jointure, how="inner"),
    dfs_a_fusionner
)

# Affichage des résultats APRÈS fusion
print("\n" + "="*70)
print("RÉSULTAT DE LA FUSION")
print("="*70)
print(f"Shape final : {df_worldbank.shape}")
print(f"   Nombre de pays (lignes) : {df_worldbank.shape[0]}")
print(f"   Nombre de colonnes : {df_worldbank.shape[1]}")

# Affichage d'un aperçu des premières lignes
print("\nAperçu des données (5 premières lignes) :")
print(df_worldbank.head())

print("\n" + "="*70)
print("✓ Fusion terminée - df_worldbank créé avec succès")
print("="*70)
```

**⚠️ IMPORTANT : Incompatibilité des codes pays FAO vs World Bank**

Les codes affichés ci-dessus sont les **codes FAO** (Food and Agriculture Organization), qui sont différents des **codes ISO 3166-1 alpha-3** utilisés par la Banque mondiale.

**Exemples de différences :**
- FAO peut utiliser des codes numériques (ex: 2, 351) ou des codes régionaux
- World Bank utilise les codes ISO standards (ex: "FRA", "USA", "CHN")

**Conséquence pour notre analyse :**

Nous ne pouvons PAS fusionner directement `df_volailles_final` et `df_worldbank` en utilisant les codes pays, car :
1. Les fichiers FAO ont les noms de pays **en français** (ex: "France", "États-Unis")
2. Les fichiers World Bank ont les noms de pays **en anglais** (ex: "France", "United States")
3. Les codes FAO ≠ codes ISO World Bank

**Solution nécessaire :**

Nous devons créer un **fichier de correspondance (mapping)** qui associe :
- `Country_FR` (nom français FAO) → `Country Name` (nom anglais World Bank) → `Country Code` (code ISO)

Cette étape de mapping sera effectuée dans les sections suivantes en utilisant l'algorithme de **fuzzy matching** (correspondance approximative) pour automatiser au maximum la correspondance des noms.

```python
# -------------------------------------------------------------------
# Vérification des valeurs manquantes
# -------------------------------------------------------------------
# .isna() : retourne True pour chaque valeur NaN, False sinon
# .sum().sum() : premier .sum() compte par colonne, deuxième .sum() fait le total
nb_nan = df_volailles_wide.isna().sum().sum()
print(f"   Valeurs manquantes : {nb_nan}")

# -------------------------------------------------------------------
# Analyse détaillée des valeurs manquantes (si présentes)
# -------------------------------------------------------------------
# Si on a trouvé au moins une valeur manquante, afficher les détails
if nb_nan > 0:
    print("\n⚠️ ANALYSE DES VALEURS MANQUANTES")
    print("   " + "-"*66)
    
    # Étape 1 : Compter les NaN dans chaque colonne
    print("\n   Répartition par colonne :")
    # MÉTHODE RAPIDE : utiliser .isna().sum() sur tout le DataFrame d'un coup
    # Cela compte les NaN dans chaque colonne en une seule opération vectorisée
    nan_par_colonne = df_volailles_wide.isna().sum()
    
    # Filtrer pour garder seulement les colonnes avec au moins un NaN
    # nan_par_colonne > 0 : crée un masque True/False
    # [masque] : garde seulement les éléments où masque = True
    colonnes_avec_nan = nan_par_colonne[nan_par_colonne > 0]
    
    # Afficher les résultats
    # .items() : parcourt (nom_colonne, nombre_nan)
    for col, count in colonnes_avec_nan.items():
        print(f"      {col} : {count} valeur(s) manquante(s)")
    
    # Étape 2 : Identifier les pays avec des valeurs manquantes
    # Construction du filtre :
    # df.isna() : DataFrame de True/False (True = NaN)
    # .any(axis=1) : vérifie s'il y a AU MOINS un True par ligne (axis=1 = horizontal)
    # df[condition] : garde seulement les lignes où condition = True
    # Résultat : DataFrame contenant uniquement les pays avec au moins un NaN
    pays_avec_nan = df_volailles_wide[df_volailles_wide.isna().any(axis=1)]
    
    print(f"\n   Pays concernés ({len(pays_avec_nan)} pays) :")
    
    # Étape 3 : Afficher le détail pour chaque pays problématique
    # MÉTHODE RAPIDE : utiliser .apply() avec une fonction lambda
    # C'est beaucoup plus rapide que .iterrows() car pandas optimise l'opération
    
    # Pour chaque pays avec NaN, construire une chaîne "Pays : colonne1, colonne2"
    # axis=1 : appliquer la fonction sur chaque ligne (horizontalement)
    # lambda row : fonction anonyme qui prend une ligne en paramètre
    # row[row.isna()] : garde seulement les colonnes où valeur = NaN
    # .index.tolist() : récupère les noms de ces colonnes
    # ', '.join() : transforme la liste en texte séparé par virgules
    details = pays_avec_nan.apply(
        lambda row: f"      • {row['Country']} : {', '.join(row[row.isna()].index.tolist())}", 
        axis=1
    )
    
    # Afficher tous les résultats
    # '\n'.join() : combine toutes les lignes avec un retour à la ligne
    print('\n'.join(details.tolist()))
    
    # Recommandation finale
    print("\n   💡 Recommandation : Ces pays devront être supprimés")
    print("      car l'analyse nécessite des données complètes sur la disponibilité de volaille")
```

```python
# ========================================================================
# Filtrage et transformation des données population
# ========================================================================
# But : créer un DataFrame avec une ligne par pays et une colonne par année
# Transformation : format long → format wide (pivot)

print("="*70)
print("FILTRAGE ET PIVOT DES DONNÉES POPULATION")
print("="*70)

# -------------------------------------------------------------------
# ÉTAPE 1 : Sélection des colonnes nécessaires
# -------------------------------------------------------------------
# On garde seulement 3 colonnes :
# - Zone : nom du pays en français (ex: "France", "États-Unis")
# - Année : année de la mesure (2000 à 2018)
# - Valeur : population en milliers de personnes

df_pop = df_Population_2000_2018[["Zone", "Année", "Valeur"]].copy()

print(f"\nÉTAPE 1 : Sélection des colonnes")
print(f"   Colonnes gardées : {df_pop.columns.tolist()}")
print(f"   Shape : {df_pop.shape}")

# -------------------------------------------------------------------
# ÉTAPE 2 : Conversion de l'unité de mesure
# -------------------------------------------------------------------
# Problème : les données sont en "milliers de personnes"
# Solution : multiplier par 1000 pour obtenir le nombre réel de personnes
# Exemple : 67 000 (milliers) × 1000 = 67 000 000 personnes

df_pop["Valeur"] = df_pop["Valeur"] * 1000

print(f"\nÉTAPE 2 : Conversion des unités")
print(f"   Milliers de personnes → Personnes (× 1000)")
```

```python
# Transformation pivot (long → wide)

print(f"\nÉTAPE 3 : Transformation pivot")
print(f"   Format AVANT pivot : {df_pop.shape} (format long)")

# .pivot() : transforme lignes en colonnes
# - index="Zone" : une ligne par pays
# - columns="Année" : chaque année devient une colonne
# - values="Valeur" : valeurs de population dans les cellules
# .reset_index() : transforme "Zone" de l'index en colonne normale

df_population_wide = (
    df_pop
    .pivot(index="Zone", columns="Année", values="Valeur")
    .reset_index()
)

print(f"   Format APRÈS pivot : {df_population_wide.shape} (format wide)")
print(f"   Colonnes créées : {df_population_wide.columns.tolist()}")

print("\nAperçu des données pivotées :")
print(df_population_wide.head())

print("\n" + "="*70)
print("✓ Filtrage et pivot terminés")
print("="*70)
```

### 4. Finalisation du DataFrame

```python
# ========================================================================
# Calcul des indicateurs population et finalisation
# ========================================================================
# But : créer 2 indicateurs clés :
# 1. Population totale en 2017 (en millions)
# 2. Croissance de la population 2017→2018 (en pourcentage)

print("="*70)
print("CALCUL DES INDICATEURS POPULATION")
print("="*70)

# -------------------------------------------------------------------
# ÉTAPE 1 : Renommage de la colonne "Zone" en "Country"
# -------------------------------------------------------------------
# Pour cohérence avec df_volailles_wide (même nom de colonne)

df_population_wide.rename(
    columns={"Zone": "Country"},
    inplace=True
)

print(f"\nÉTAPE 1 : Renommage")
print(f"   'Zone' → 'Country' (pour fusion future)")

# -------------------------------------------------------------------
# ÉTAPE 2 : Création de l'indicateur "Population total 2017 (millions)"
# -------------------------------------------------------------------
# Extraire la population de l'année 2017 et convertir en millions
# Division par 1 000 000 pour passer de personnes → millions de personnes
# Arrondi à 1 décimale pour plus de lisibilité
# Cet indicateur représente la taille du marché potentiel

df_population_wide["Population total 2017 (millions)"] = (
    df_population_wide[2017] / 1_000_000
).round(1)

print(f"\nÉTAPE 2 : Population totale 2017")
print(f"   Colonne créée : 'Population total 2017 (millions)'")
print(f"   Conversion : personnes → millions (÷ 1 000 000)")
print(f"   Arrondi : 1 décimale")
print(f"   Exemple : {df_population_wide['Population total 2017 (millions)'].iloc[0]} millions")

# -------------------------------------------------------------------
# ÉTAPE 3 : Calcul de la croissance de population 2017→2018 (%)
# -------------------------------------------------------------------
# Formule : ((Population_2018 - Population_2017) / Population_2017) × 100
# Résultat : taux de croissance en pourcentage
# Positif = croissance, Négatif = déclin
# Arrondi à 2 décimales pour précision
# Cet indicateur mesure la dynamique démographique récente

df_population_wide["Population growth 2017-2018 (%)"] = (
    ((df_population_wide[2018] - df_population_wide[2017]) / df_population_wide[2017]) * 100
).round(2)

print(f"\nÉTAPE 3 : Croissance population 2017-2018")
print(f"   Colonne créée : 'Population growth 2017-2018 (%)'")
print(f"   Formule : ((Pop_2018 - Pop_2017) / Pop_2017) × 100")
print(f"   Arrondi : 2 décimales")
print(f"   Exemple : {df_population_wide['Population growth 2017-2018 (%)'].iloc[0]}%")

# -------------------------------------------------------------------
# ÉTAPE 4 : Sélection des colonnes finales
# -------------------------------------------------------------------
# On garde seulement les 3 colonnes nécessaires :
# - Country : nom du pays en français (pour fusion avec df_volailles_wide)
# - Population total 2017 (millions) : taille du marché
# - Population growth 2017-2018 (%) : dynamique démographique

df_population_final = df_population_wide[
    [
        "Country",                              # Nom français du pays
        "Population total 2017 (millions)",     # Population en millions
        "Population growth 2017-2018 (%)"       # Croissance en %
    ]
].copy()

print(f"\nÉTAPE 4 : Sélection des colonnes finales")
print(f"   Colonnes gardées : {df_population_final.columns.tolist()}")

# -------------------------------------------------------------------
# Vérification du résultat final
# -------------------------------------------------------------------

print(f"\nRÉSULTAT FINAL")
print(f"   Shape : {df_population_final.shape}")
print(f"   Nombre de pays : {len(df_population_final)}")
```

```python
# -------------------------------------------------------------------
# Vérification des valeurs manquantes
# -------------------------------------------------------------------
# .isna() : retourne True pour chaque valeur NaN, False sinon
# .sum().sum() : premier .sum() compte par colonne, deuxième .sum() fait le total
nb_nan = df_population_final.isna().sum().sum()
print(f"   Valeurs manquantes : {nb_nan}")

# -------------------------------------------------------------------
# Analyse détaillée des valeurs manquantes (si présentes)
# -------------------------------------------------------------------
# Si on a trouvé au moins une valeur manquante, afficher les détails
if nb_nan > 0:
    print("\n⚠️ ANALYSE DES VALEURS MANQUANTES")
    print("   " + "-"*66)
    
    # Étape 1 : Compter les NaN dans chaque colonne
    print("\n   Répartition par colonne :")
    # MÉTHODE RAPIDE : utiliser .isna().sum() sur tout le DataFrame d'un coup
    # Cela compte les NaN dans chaque colonne en une seule opération vectorisée
    nan_par_colonne = df_population_final.isna().sum()
    
    # Filtrer pour garder seulement les colonnes avec au moins un NaN
    # nan_par_colonne > 0 : crée un masque True/False
    # [masque] : garde seulement les éléments où masque = True
    colonnes_avec_nan = nan_par_colonne[nan_par_colonne > 0]
    
    # Afficher les résultats
    # .items() : parcourt (nom_colonne, nombre_nan)
    for col, count in colonnes_avec_nan.items():
        print(f"      {col} : {count} valeur(s) manquante(s)")
    
    # Étape 2 : Identifier les pays avec des valeurs manquantes
    # Construction du filtre :
    # df.isna() : DataFrame de True/False (True = NaN)
    # .any(axis=1) : vérifie s'il y a AU MOINS un True par ligne (axis=1 = horizontal)
    # df[condition] : garde seulement les lignes où condition = True
    # Résultat : DataFrame contenant uniquement les pays avec au moins un NaN
    pays_avec_nan = df_population_final[df_population_final.isna().any(axis=1)]
    
    # MÉTHODE RAPIDE : utiliser .apply() avec une fonction lambda
    # C'est beaucoup plus rapide que .iterrows() car pandas optimise l'opération
    
    # Pour chaque pays avec NaN, construire une chaîne "Pays : colonne1, colonne2"
    # axis=1 : appliquer la fonction sur chaque ligne (horizontalement)
    # lambda row : fonction anonyme qui prend une ligne en paramètre
    # row[row.isna()] : garde seulement les colonnes où valeur = NaN
    # .index.tolist() : récupère les noms de ces colonnes
    # ', '.join() : transforme la liste en texte séparé par virgules
    details = pays_avec_nan.apply(
        lambda row: f"      • {row['Country']} : {', '.join(row[row.isna()].index.tolist())}", 
        axis=1
    )
    
    # Afficher tous les résultats
    # '\n'.join() : combine toutes les lignes avec un retour à la ligne
    print('\n'.join(details.tolist()))
    
    # Recommandation finale
    print("\n   💡 Recommandation : Ces pays devront être supprimés")
    print("      car l'analyse nécessite des données complètes sur la population")
```

## 3. Creation de mapping

### 1. Сomparer les noms français des pays avec les noms anglais standardisés de la Banque mondiale

```python
# ========================================================================
# Création automatique du mapping FR → EN avec fuzzy matching
# ========================================================================
# But : associer automatiquement les noms français (FAO) aux noms anglais (World Bank)
# Méthode : algorithme de correspondance approximative (fuzzy matching) avec rapidfuzz
# Résultat : fichier CSV de correspondance à vérifier manuellement

from rapidfuzz import process, fuzz
import pandas as pd
import numpy as np

print("="*70)
print("CRÉATION DU MAPPING PAYS FR → EN (Fuzzy Matching)")
print("="*70)

# ===================================================================
# ÉTAPE 1 : Extraction des listes de pays des deux sources
# ===================================================================

# Liste des pays en FRANÇAIS (depuis df_volailles_wide - données FAO)
# sorted() : trie par ordre alphabétique pour faciliter la vérification
# .unique() : garde seulement les valeurs uniques (pas de doublons)
pays_fr = sorted(df_volailles_wide["Country"].unique())

# Liste des pays en ANGLAIS (depuis df_worldbank - données World Bank)
pays_en = sorted(df_worldbank["Country Name"].unique())

print(f"\nNombre de pays FAO (français) : {len(pays_fr)}")
print(f"Nombre de pays World Bank (anglais) : {len(pays_en)}")

# ===================================================================
# ÉTAPE 2 : Définition de la fonction de matching
# ===================================================================

def matcher_pays(nom_fr):
    """    
    Trouve le meilleur nom anglais correspondant à un nom français.
    
    Paramètre:
        nom_fr : nom du pays en français (ex: "États-Unis")
    
    Retourne:
        Series pandas avec 4 colonnes :
        - Country_FR : nom français original
        - Country_EN : meilleur match en anglais
        - statut : "OK" si score ≥ 95, sinon "A_VERIFIER"
        - score : score de similarité (0-100)
    
    Méthode:
        Utilise fuzz.WRatio (Weighted Ratio) qui :
        - Compare les chaînes de caractères
        - Ignore la casse (majuscules/minuscules)
        - Tolère les différences d'accents et de ponctuation
        - Donne un score de 0 (pas de similarité) à 100 (identique)
    """
    
    # process.extractOne() : trouve la meilleure correspondance
    # Retourne : (meilleur_match, score, index)
    match, score, _ = process.extractOne(
        nom_fr,                    # texte à chercher
        pays_en,                   # liste où chercher
        scorer=fuzz.WRatio         # algorithme de calcul de similarité
    )
    
    # Définir le statut selon le score de confiance
    # Score ≥ 95 : très haute confiance → OK
    # Score < 95 : vérification manuelle nécessaire → A_VERIFIER
    statut = "OK" if score >= 95 else "A_VERIFIER"
    
    # Retourner les résultats sous forme de Series pandas
    return pd.Series({
        "Country_FR": nom_fr,      # nom français original
        "Country_EN": match,       # meilleur match anglais trouvé
        "statut": statut,          # statut de validation
        "score": score             # score de similarité (pour analyse)
    })

print("\n✓ Fonction de matching définie")

# ===================================================================
# ÉTAPE 3 : Application du matching sur tous les pays
# ===================================================================

print("\nApplication du fuzzy matching sur tous les pays...")

# List comprehension : [matcher_pays(n) for n in pays_fr]
# Applique la fonction matcher_pays() à chaque pays français
# Résultat : liste de Series pandas
# pd.DataFrame() : convertit la liste de Series en DataFrame
df_correspondance = pd.DataFrame([matcher_pays(n) for n in pays_fr])

print(f"✓ Matching terminé : {len(df_correspondance)} correspondances créées")

# ===================================================================
# ÉTAPE 4 : Analyse des résultats
# ===================================================================

# Compter combien de correspondances sont "OK" vs "A_VERIFIER"
nb_ok = (df_correspondance["statut"] == "OK").sum()
nb_a_verifier = (df_correspondance["statut"] == "A_VERIFIER").sum()

print(f"\nRésultats du matching :")
print(f"   Correspondances OK (score ≥ 95) : {nb_ok}")
print(f"   À vérifier manuellement (score < 95) : {nb_a_verifier}")

# Afficher les cas à vérifier (si présents)
if nb_a_verifier > 0:
    print("\n⚠️ Pays nécessitant une vérification manuelle :")
    pays_a_verifier = df_correspondance[df_correspondance["statut"] == "A_VERIFIER"]
    for idx, row in pays_a_verifier.iterrows():
        print(f"   {row['Country_FR']} → {row['Country_EN']} (score: {row['score']})")

# ===================================================================
# ÉTAPE 5 : Préparation du fichier de correspondance
# ===================================================================

# Garder seulement les colonnes nécessaires pour l'export
# La colonne "score" est utile pendant l'analyse mais pas nécessaire dans le fichier final
df_correspondance_finale = df_correspondance[["Country_FR", "Country_EN", "statut"]].copy()

print("\n" + "="*70)
print("APERÇU DU MAPPING CRÉÉ")
print("="*70)
print(df_correspondance_finale.head(10))

print("\n✓ Mapping automatique créé avec succès")
print("📌 Étape suivante : sauvegarder en CSV pour vérification manuelle")
```

# IV. Création de l'ensemble de données final

### 1. Création et aperçu du dataframe final

```python
# ========================================================================
# CRÉATION DU DATASET FINAL df_poulet
# ========================================================================
# Objectif : Fusionner 4 DataFrames en conservant toutes les lignes de df_volailles_wide
# Ordre des fusions :
# 1. df_volailles_wide (BASE) + df_population_final
# 2. Résultat + df_mapping (pont FR → EN)
# 3. Résultat + df_worldbank (données économiques)

print("="*70)
print("FUSION DES DATAFRAMES - CRÉATION DE df_poulet")
print("="*70)

# ========================================================================
# ÉTAPE 1 : df_volailles_clean + df_population_final
# ========================================================================
# Jointure par le nom du pays (en français)
# how='left' : garde TOUTES les lignes de df_volailles_clean

print("\n📊 ÉTAPE 1 : Ajout des données de population")
print(f"   df_volailles_wide : {df_volailles_wide.shape}")
print(f"   df_population_final : {df_population_final.shape}")

df_step1 = pd.merge(
    df_volailles_wide,            # DataFrame de BASE (gauche)
    df_population_final,          # DataFrame à ajouter (droite)
    left_on='Country',            # Colonne de jointure à gauche (nom français)
    right_on='Country',           # Colonne de jointure à droite
    how='left'                    # LEFT JOIN : garde toutes les lignes de gauche
)

print(f"   Résultat après fusion : {df_step1.shape}")
print(f"   ✓ Population ajoutée")

# ========================================================================
# ÉTAPE 2 : Résultat + df_mapping
# ========================================================================
# Jointure par le nom français pour obtenir le nom anglais et le code pays
# df_mapping contient : Country_FR, Country_EN, Code Country

print("\n📊 ÉTAPE 2 : Ajout du mapping FR → EN")
print(f"   df_step1 : {df_step1.shape}")
print(f"   df_mapping : {df_mapping.shape}")

df_step2 = pd.merge(
    df_step1,                     # DataFrame de l'étape précédente
    df_mapping,                   # DataFrame de mapping
    left_on='Country',            # Nom français dans df_step1
    right_on='Country_FR',        # Nom français dans df_mapping
    how='left'                    # LEFT JOIN : garde toutes les lignes
)

print(f"   Résultat après fusion : {df_step2.shape}")
print(f"   ✓ Noms anglais et codes pays ajoutés")

# ========================================================================
# ÉTAPE 3 : Résultat + df_worldbank
# ========================================================================
# Jointure par le nom anglais ET le code pays pour plus de fiabilité
# df_worldbank contient : Country Name, Country Code + indicateurs économiques

print("\n📊 ÉTAPE 3 : Ajout des données de la World Bank")
print(f"   df_step2 : {df_step2.shape}")
print(f"   df_worldbank : {df_worldbank.shape}")

df_poulet = pd.merge(
    df_step2,                     # DataFrame de l'étape précédente
    df_worldbank,                 # DataFrame World Bank
    left_on=['Country_EN', 'Code Country'],      # Colonnes de jointure à gauche
    right_on=['Country Name', 'Country Code'],   # Colonnes de jointure à droite
    how='left'                    # LEFT JOIN : garde toutes les lignes
)

print(f"   Résultat final : {df_poulet.shape}")
print(f"   ✓ Indicateurs économiques ajoutés")
```

```python
# ========================================================================
# DÉTECTION COMPLÈTE DES VALEURS MANQUANTES (NaN)
# ========================================================================

print("="*70)
print("ANALYSE COMPLÈTE DES VALEURS MANQUANTES")
print("="*70)

# ========================================================================
# 1. Statistiques générales par colonne
# ========================================================================
print("\n📊 Nombre de NaN par colonne :")
print("-"*70)
nan_par_colonne = df_poulet.isna().sum()
nan_par_colonne_pct = (df_poulet.isna().sum() / len(df_poulet) * 100).round(2)

# Afficher seulement les colonnes avec au moins 1 NaN
colonnes_avec_nan = nan_par_colonne[nan_par_colonne > 0]
if len(colonnes_avec_nan) > 0:
    for col in colonnes_avec_nan.index:
        print(f"   {col:50s} : {nan_par_colonne[col]:3d} NaN ({nan_par_colonne_pct[col]:5.2f}%)")
else:
    print("   ✓ Aucune valeur manquante détectée !")

# ========================================================================
# 2. Identification des pays avec au moins 1 NaN
# ========================================================================
print("\n" + "="*70)
print("LISTE DES PAYS AVEC DES VALEURS MANQUANTES")
print("="*70)

# Trouver toutes les lignes contenant au moins 1 NaN
lignes_avec_nan = df_poulet[df_poulet.isna().any(axis=1)]

if len(lignes_avec_nan) > 0:
    print(f"\n📍 Nombre total de pays avec NaN : {len(lignes_avec_nan)}")
    print(f"   (soit {len(lignes_avec_nan)/len(df_poulet)*100:.2f}% du dataset)")
    print("\n" + "-"*70)
    
    # Pour chaque pays avec NaN, afficher les colonnes concernées
    for idx, row in lignes_avec_nan.iterrows():
        pays = row['Country']
        # Trouver quelles colonnes ont NaN pour ce pays
        colonnes_nan = row[row.isna()].index.tolist()
        
        print(f"\n🔸 {pays}")
        print(f"   Colonnes avec NaN ({len(colonnes_nan)}) : {', '.join(colonnes_nan)}")
    
    print("\n" + "="*70)
    print(f"RÉSUMÉ : {len(lignes_avec_nan)} pays ont des valeurs manquantes")
    print("="*70)
else:
    print("\n✅ Aucun pays avec des valeurs manquantes !")
    print("="*70)
```
